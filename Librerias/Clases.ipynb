## Compuerta de Hadamard $$ H $$

Actua sobre solo un qbit

$$H=\begin{bmatrix}  \frac{1}{\sqrt2} & \frac{1}{\sqrt2} \\
       \frac{1}{\sqrt2} & \frac{-1}{\sqrt2} \\\end{bmatrix}$$

$$ \begin{bmatrix}  \frac{1}{\sqrt2} \\ \frac{1}{\sqrt2} \end{bmatrix} = {\displaystyle | + \rangle }  $$

$$ \begin{bmatrix}  \frac{1}{\sqrt2} \\ \frac{-1}{\sqrt2} \end{bmatrix} = {\displaystyle | - \rangle }  $$


$$ H  {\displaystyle | 0 \rangle }  = {\displaystyle | +\rangle }  $$       
       

       
$$ H {\displaystyle | 1 \rangle }= {\displaystyle | - \rangle }  $$




$$ H  {\displaystyle | +\rangle } =  \begin{bmatrix}  1 \\ 0 \end{bmatrix} =  {\displaystyle | 0 \rangle }  $$

# Ejercicio Ciudades

Teniendo en cuenta que el angulo de latitud no se esta midiendo de la misma forma que las coordenadas esfericas se procede a hacer la operacion 90-L y pasarlos a radianes

<img src="Imagenes/Bloch_sphere.png" width=30%>

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook 
phi, theta = np.mgrid[0 : 2 * np.pi : 0.05, 0 : np.pi : 0.05]  
x = np.cos(phi) * np.sin(theta)
y = np.sin(phi) * np.sin(theta)
z = np.cos(theta)
plt.rcParams.update({'font.size': 5})
Ciudades = [
    ['Londres', 51.5, -0.1167],
    ['Bogotá', 4.5964, -74.0833],
    ['Roma', 41.8960, 12.4833],
    ['Melbourne', -37.8200, 144.9750],
    ['Nairobi', -1.2833, 36.8167],
    ['Santiago', -33.4500, -70.6670],
    ['Nueva York', 40.6943, -73.9249],
    ['Los Angeles', 34.1139, -118.4068],
    ['Moscú', 55.7522, 37.6155],
    ['Quito', -0.2150, -78.5001],
    ['Tokio', 35.6850, 139.7514],
    ['Lagos', 6.4433, 3.3915],
    ['Atenas', 37.9833, 23.7333],
    ['Belén de Pará', -1.4500, -48.4800],
    ['Budapest', 47.5000, 19.0833],
    ['Vancouver', 49.2734, -123.1216],
    ['París', 48.8667, 2.3333],
    ['Puebla', 19.0500, -98.2000]
]
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(x, y, z, color='orange', rstride=2, cstride=2, linewidth=1, alpha=0.5)
for i in Ciudades:    
    lonr=i[2]*np.pi/180 
    latr=(90-i[1])*np.pi/180  
    x = np.cos(lonr)*np.sin(latr)
    y = np.sin(lonr)*np.sin(latr)
    z = np.cos(latr)
    ax.scatter(x, y, z, color='red', s=10) 
    ax.text(x, y, z, i[0], color='black', )   
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_zlabel('Eje Z')
plt.show()
```
