```python
from libreriacomplejos import complejo as C
import numpy as np
import matplotlib.pyplot as plt

def mandelbrot(c):
    z=C(0,0)  
    color=0
    ans=True
    while(ans and color<30):
        z=C.suma(C.potencia(z,2),c)     
        if(z.r>2):
            ans=False
        color=color+1
    return ans, color
def graf():
    plano=np.zeros((400,700))
    for parte_real in range(700):
        for parte_img in range(400):
            num=C(parte_real/200-2.5,parte_img/200-1)
            p,color=mandelbrot(num)
            if(p):
                plano[parte_img][parte_real]=0
            else:
                plano[parte_img][parte_real]=color
    plt.figure(figsize=(10,10))
    plt.imshow(plano,'magma')
    plt.show()
```

```python
print(mandelbrot(complejo(-0.1,1.1)))
```

```python
graf()
```
